```python
def play_game(env_4_connect,player_0,player_1):
    env_4_connect.reset()

    ### First Observation
    observations = env_4_connect.last()

    while not (observations[2]) and not (observations[3]):
        # Player 0 play
        action = player_0.get_value(observations)
        env_4_connect.step(action)


        # Observation Player 1
        observations = env_4_connect.last()


        if not (observations[2]):
            action = player_1.get_value(observations)
            env_4_connect.step(action)


        # Observation Player 0
        observations = env_4_connect.last()

    ### Affect the reward
    game_rewards = env_4_connect.rewards
    return (np.array(list(game_rewards.values()))>0).astype(int)
```

```python
def play_n_games(env_4_connect,player_0,player_1,n):
    rep = np.zeros(2)
    for i in range(n):
        if i%2 == 0 :
            rep+=play_game(env_4_connect,player_0,player_1)
        else:
            rep+=play_game(env_4_connect,player_1,player_0)[::-1]
    return rep
```
